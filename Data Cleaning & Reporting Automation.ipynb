```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "CustomerID":[101,102,103,104,101,105],
    "Name":["Ravi","Sita","Rahul","Anita","Ravi","Kiran"],
    "Age":[25,None,30,28,25,35],
    "City":["Hyderabad","Hyderabad","Delhi","Mumbai","Hyderabad","Bangalore"],
    "Sales":[500,700,None,600,500,900]
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv("sales_data.csv", index=False)

print("/content/sales_data.csv")
```

```python
data = pd.read_csv("sales_data.csv")

print("Original Data")
data
```

```python
data.isnull().sum()
```

```python
data["Age"].fillna(data["Age"].mean(), inplace=True)
data["Sales"].fillna(data["Sales"].median(), inplace=True)

data
```

```python
data = data.drop_duplicates()

data
```

```python
data["City"] = data["City"].str.strip().str.title()

data
```

```python
data.to_csv("cleaned_sales_data.csv", index=False)

print("Cleaned data saved")
```

```python
total_sales = data["Sales"].sum()
avg_sales = data["Sales"].mean()
avg_age = data["Age"].mean()

print("Total Sales:", total_sales)
print("Average Sales:", avg_sales)
print("Average Age:", avg_age)
```

```python
city_sales = data.groupby("City")["Sales"].sum()

city_sales
```

```python
plt.figure()

city_sales.plot(kind="bar")

plt.title("Sales by City")
plt.xlabel("City")
plt.ylabel("Sales")

plt.show()
```
